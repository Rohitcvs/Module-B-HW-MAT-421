Section 1: Interpolation Problem Statement
Interpolation is a fundamental mathematical technique used to estimate the values of a function for points that lie within the range of a discrete set of known data points. The goal is to construct new data points by "filling in the gaps" between the given values, based on the assumption that the data varies smoothly between the known points. Interpolation is widely used in various fields, such as engineering, finance, computer graphics, and scientific research, where obtaining data at every desired point is impractical or impossible.

The process of interpolation relies on constructing a function, either linear or nonlinear, that passes through the known points and smoothly transitions between them. This function can take various forms depending on the interpolation method used, such as linear interpolation, polynomial interpolation, or spline interpolation. The choice of method depends on the nature of the data and the desired level of accuracy.


```python
import numpy as np
import matplotlib.pyplot as plt

# Known data points
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 1, 4, 9, 16, 25])

# Visualize the data points
plt.scatter(x, y, color='red', label='Known Data Points')
plt.title("Interpolation Problem")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

Section 2: Linear Interpolation
Linear interpolation is widely used because of its simplicity and efficiency. It works particularly well for datasets where the variation between points is small or approximately linear. For example, it can be used to estimate temperatures between recorded times of the day, predict financial trends between two trading periods, or fill in missing values in datasets. Its computational efficiency makes it suitable for real-time applications, especially when large amounts of data need to be processed quickly.

However, linear interpolation has its limitations. It assumes that the data follows a straight-line pattern, which may not accurately reflect the actual trend if the data is nonlinear. It also does not handle extrapolation well, as predicting values outside the range of the known points can lead to significant inaccuracies. Additionally, the method produces discontinuities in the derivative of the interpolated function, as each interval between data points is treated independently.

```python
from scipy.interpolate import interp1d

# Linear interpolation function
linear_interp = interp1d(x, y, kind='linear')

# Generate interpolated values
x_new = np.linspace(0, 5, 50)
y_new = linear_interp(x_new)

# Plot
plt.plot(x, y, 'o', label='Known Points')
plt.plot(x_new, y_new, '-', label='Linear Interpolation')
plt.title("Linear Interpolation")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

Section 3: Cubic Spline Interpolation
Cubic spline interpolation is a mathematical method used to create a smooth curve that passes through a set of known data points. Unlike linear interpolation, which connects points with straight lines, 
or polynomial interpolation, which uses a single high-degree polynomial, cubic spline interpolation constructs a piecewise cubic polynomial for each segment between data points. These cubic polynomials 
are designed to join smoothly, ensuring both the function and its first and second derivatives are continuous across the entire range of interpolation.

One of the key advantages of cubic spline interpolation is its ability to produce smooth and visually appealing curves, making it an essential tool in 
fields like computer graphics, animation, and data visualization. It is also widely used in engineering and scientific applications, such as modeling 
the trajectory of moving objects, analyzing experimental data, or interpolating values in physical simulations.

```python
from scipy.interpolate import CubicSpline

# Cubic spline interpolation function
cubic_spline = CubicSpline(x, y)

# Generate interpolated values
y_cubic = cubic_spline(x_new)

# Plot
plt.plot(x, y, 'o', label='Known Points')
plt.plot(x_new, y_cubic, '-', label='Cubic Spline Interpolation')
plt.title("Cubic Spline Interpolation")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
